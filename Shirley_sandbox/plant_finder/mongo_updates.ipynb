```python
# pull in our data
# clean our data
# save cleaned data to db
# send data to browser using flask
# build a chart
import os
import pandas as pd
from pymongo import MongoClient

client = MongoClient(os.getenv("MONGO_URI", "mongodb://localhost:27017/"))
db = client.flowers
```

```python
file_one = "./flowers.csv"
flowers = pd.read_csv(file_one)
drop_na = flowers.dropna()
cleaned = drop_na[['Common_Name', 'Scientific_Name', 'Family_Common_Name', 'Duration', 
        'Flower_Color', 'Foliage_Color', 'Growth_Habit',
        'Height,_Mature_(feet)', 
        'Toxicity', 'Drought_Tolerance',
       'Shade_Tolerance', 'Bloom_Period', 
       'Palatable_Human']]
cleaned
```

```python
# create dictionary from dataframe for mongo
flower_dict = cleaned.to_dict("records")
flower_dict
```

```python
collection = db['flowers']
```

```python
collection.insert_many(flower_dict)
```
