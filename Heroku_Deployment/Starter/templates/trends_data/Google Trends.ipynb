```python
from pytrends.request import TrendReq

%matplotlib inline

import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
#connects to 
googleConn = TrendReq(hl='en-US',
                    tz=360,
                    timeout=(10,25),
                    retries=2,
                    backoff_factor=0.1)
```

```python
plant_search = ['plants']  
#creates a list with up to 5 search terms
googleConn.build_payload(plant_search)
df = googleConn.interest_over_time()
```

```python
df.head()
```

```python
df.plot()
```

```python
#Dataframe comparing tree vs flowers by each of the states

states_data = googleConn.interest_by_region()
states_data
```

```python
states_top = states_data.loc[states_data.trees > states_data.trees.quantile(0.75)]
```

```python
#display of bar graph of the top states for tree search data
states_top.plot.barh()
```

```python
states_top.to_csv('states_top.csv', encoding='utf-8')
states_top.to_json('states_top.json')

googleConn.build_payload(['trees', 'flowers', 'houseplants'], cat=5)
```
