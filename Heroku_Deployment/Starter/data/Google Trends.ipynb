```python
from pytrends.request import TrendReq

%matplotlib inline

import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
#connects to 
googleConn = TrendReq(hl='en-US',
                    tz=360,
                    timeout=(10,25),
                    retries=2,
                    backoff_factor=0.1)
```

```python
plant_search = ['plants']  
#creates a list with up to 5 search terms
googleConn.build_payload(plant_search)
df = googleConn.interest_over_time()
```
