```python
from bs4 import BeautifulSoup
import pymongo
import pandas as pd
import requests
from sqlalchemy import create_engine
import json
```

```python
Northern_America = [601,602,603,604,605,606]
```

```python
state_slug = []
state_name = []
zone_number = []

for zone in Northern_America:
    url = f"https://trefle.io/api/v1/distributions/{zone}?token=lJB461ruHXGiw6SBUD_liGd6XLjQI6Kyqj0AGLhhKn0"
    print("downloading: " + url)
    response = requests.get(url)
    data = response.json()['data']['children']
    
    for state in data:
        state_slug.append(state['slug'])
        state_name.append(state['name'])
        zone = zone
        zone_number.append(zone)
        
```

```python
zone_name = []
zone_id = []
for zone in Northern_America:
    url = f"https://trefle.io/api/v1/distributions/{zone}?token=lJB461ruHXGiw6SBUD_liGd6XLjQI6Kyqj0AGLhhKn0"
    print("downloading: " + url)
    response = requests.get(url)
    data = response.json()['data']['name']
    zone = zone
    zone_id.append(zone)
    zone_name.append(data)
    
    
```

```python
zone_info = pd.DataFrame({"zone_id": zone_id, "zone_name": zone_name})
zone_info
```

```python
import sqlite3
```

```python
conn = sqlite3.connect('East_us.db')

conn.cursor().execute("create table if not exists east_species ({})".format(','.join(Est_us.columns)))
```

```python
for row in Est_us.iterrows():
    sql = 'insert into east_species ({}) values ({})'.format(','.join(Est_us.columns),','.join(['?']*len(Est_us.columns)))
    conn.cursor().execute(sql,tuple(row[1]))
    
conn.commit()
```

```python
conn = sqlite3.connect('West_us.db')

conn.cursor().execute("create table if not exists west_species ({})".format(','.join(Wst_us.columns)))
```
