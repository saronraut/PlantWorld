```python
slug_name = []
edbl = []
flwr_clr = []
grw_mean = []
toxicity = []
grw_rate = []
bloom_mnths = []
for species in species_clean:
    url = f'https://trefle.io/api/v1/species/{species}?token=lJB461ruHXGiw6SBUD_liGd6XLjQI6Kyqj0AGLhhKn0'
    response = requests.get(url)
    print(url)
    data = response.json()['data']
    slug = data['slug']
    edible = data['edible']
    flower_clr = data['flower']['color']
    grwth_avg = data['specifications']['average_height']['cm']
    toxic = data['specifications']['toxicity']
    grwth_rate = data['specifications']['growth_rate']
    bloom_months = data['growth']['bloom_months']
    
    slug_name.append(slug)
    edbl.append(edible)
    flwr_clr.append(flower_clr)
    grw_mean.append(grwth_avg)
    toxicity.append(toxic)
    grw_rate.append(grwth_rate)
    bloom_mnths.append(bloom_months)
    
    
```
